```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('sunglasses.csv')
```

```python
data
```

```python
#data['price'] = data['price'].str.replace('$','')
data['price'] = data['price'].astype('float')
```

```python
#data['list_price'] = data['list_price'].str.replace('$','')
data['list_price'] = data['list_price'].astype('float')
```

```python
#data['pct_off_list'] = data['pct_off_list'].str.replace('% off','')
data['pct_off_list'] = data['pct_off_list'].astype('float')
```

```python
data.groupby('polarized')[['price', 'list_price', 'pct_off_list']].mean()
```

```python
data.plot(style = 'hist', x = 'price')
```
