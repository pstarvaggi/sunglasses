```python
woot = pd.read_csv('woot_sunglasses.csv')
```

```python
woot
```

```python
def extract_name(x):
    try:
        return re.search('[a-zA-Z]+\d+\w*',x).group().lower()
    except:
        return None
```

```python
list(map(extract_name,woot.title))
```

```python
key_ = 'oo4124'
df.loc[list(map(lambda x: key_ in x.lower(), df.name))]
```

```python
woot.loc[list(map(lambda x: key_ in x.lower(), woot.title))]
```

```python
woot['key_'] = list(map(extract_name,woot.title))
df['key_'] = list(map(extract_name,df.name))
df_woot = pd.merge(df.loc[map(lambda x: bool(x), df.key_)], woot.loc[map(lambda x: bool(x), woot.key_)], on = 'key_', suffixes = ('_hut','_woot'))
```

```python
df_woot[['name','price_hut', 'price_woot','url_hut', 'url_woot']].sort_values(by = 'url_woot')
```

```python
import plotly
# connected=True means it will download the latest version of plotly javascript library.
plotly.offline.init_notebook_mode(connected=True)
```

```python
from plotly.graph_objs import Bar, Figure, Histogram, Layout

df_price_brands = df.groupby('brand')['price'].median().sort_values(ascending=False).reset_index()

data = [Bar(x=df_price_brands['brand'], y=df_price_brands['price'], text=df_price_brands['brand'])]
layout = Layout(title="Brands by Median Price")

fig = Figure(data=data, layout=layout)

plotly.offline.iplot(fig,show_link=False)
```

```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
```

```python
descriptions = df.description.unique()
sunglasses_mask = np.array(Image.open('sunglasses.png'))
stopwords = set(STOPWORDS)
stopwords.add('ray')
stopwords.add('ban')
stopwords.add('ray-ban')
stopwords.add('come')
```

```python
wc = WordCloud(background_color="white", max_words=2000, width=800, height=400)
wc.generate(' '.join(descriptions[1:]).lower())
```

```python
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
plt.figure(figsize = (12,6))
plt.imshow(sunglasses_mask, cmap=plt.cm.gray, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```
